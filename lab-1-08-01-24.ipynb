## To get the frequency of the words 

```python
str2 = "It looks like you never confirmed your email address, by clicking on the link in the email that was sent to you when you registered.  You must do that to finish the registration, before logging in to use the corpora. Click here to have the code re-sent to you."
# A datastructure to strore the frequencies
freq = {}
# for splitting the string by words
str2s = str2.split(" ")
for i in str2s:
    if i in freq.keys():
        freq[i] = freq[i]+1
    else:
        freq[i] = 1
```

```python
freq
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (30, 10)
plt.bar(list(freq.keys()), list(freq.values()))
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Word Frequency')
plt.show()
```
